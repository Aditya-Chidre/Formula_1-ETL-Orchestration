```python
source_file = f"{landing_folder_path}/constructors.json"
table_name = f"{catalog_name}.{bronze_schema}.constructors"
```

```python
constructors_schema = """constructorId STRING, 
                         name STRING, 
                         nationality STRING, 
                         url STRING
                         """
```

```python
constructors_df = (
    spark.read
        .format('json')
        .schema(constructors_schema)
        .option('mode','FAILFAST')
        .load(source_file)
)
```

```python
constructors_df = add_metadata_columns(constructors_df)
display(constructors_df.head(5))
```

```python
constructors_df.write.format('delta').mode('overwrite').saveAsTable(table_name)
```
