```python
target_table = f"{catalog_name}.{gold_schema}.dim_races"
races_table = f"{catalog_name}.{silver_schema}.races"
circuits_table = f"{catalog_name}.{silver_schema}.circuits"
```

```python
races_df = spark.table(races_table)
circuits_df = spark.table(circuits_table)
```

```python
dim_races_df = (
    races_df
        .join(circuits_df, races_df.circuit_id == circuits_df.circuit_id, "inner")
        .select(
            races_df.season,
            races_df.round,
            races_df.race_name,
            races_df.race_date,
            circuits_df.circuit_name,
            circuits_df.locality,
            circuits_df.country
        )
)
```

```python
dim_races_df.write.format('delta').mode('overwrite').saveAsTable(target_table)
```
